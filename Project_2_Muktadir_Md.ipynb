```python
import pandas as pd

# Read SPSS file
df = pd.read_spss("Sandwich.sav")

# Quick look
print(df.head())
print(df.info())
print("Shape:", df.shape)
print("\nColumn names:\n", df.columns)
```

```python
# Change display limits
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)
pd.set_option('display.colheader_justify', 'left')

# Show the full dataset
df
```

```python
# Number of rows and columns
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
# Count missing values per column
missing_values = df.isnull().sum()

print("\nMissing values per column:")
print(missing_values)

# Check if any missing values exist at all
if missing_values.sum() == 0:
    print("\n✅ No missing values found in the dataset.")
else:
    print("\n⚠️ There are missing values in the dataset.")
```

```python
df["Brot"] = df["Brot"].astype("category")
df["Belag"] = df["Belag"].astype("category")
df["Butter"] = df["Butter"].astype("category")
```

```python
# --- DESCRIPTIVE ANALYSIS ---

print("\n=== DESCRIPTIVE STATISTICS FOR NUMBER OF ANTS (Ameisen) ===")
desc = df["Ameisen"].describe(percentiles=[0.25, 0.5, 0.75])
print(desc)

# Rename for readability
print("\nSummary in words:")
print(f"- Count (N): {desc['count']:.0f}")
print(f"- Mean: {desc['mean']:.2f}")
print(f"- Median (50%): {desc['50%']:.2f}")
print(f"- Standard deviation: {desc['std']:.2f}")
print(f"- Minimum: {desc['min']:.2f}")
print(f"- 25th percentile (Q1): {desc['25%']:.2f}")
print(f"- 75th percentile (Q3): {desc['75%']:.2f}")
print(f"- Maximum: {desc['max']:.2f}")

# --- GROUP-WISE MEANS ---
print("\n=== MEAN NUMBER OF ANTS BY CATEGORY ===")

print("\nBy Bread Type (Brot):")
print(df.groupby("Brot", observed=True)["Ameisen"].mean())

print("\nBy Topping (Belag):")
print(df.groupby("Belag", observed=True)["Ameisen"].mean())

print("\nBy Butter Presence (Butter):")
print(df.groupby("Butter", observed=True)["Ameisen"].mean())

# --- OPTIONAL: STANDARD DEVIATION PER GROUP ---
print("\n=== STANDARD DEVIATION BY CATEGORY ===")

print("\nBy Bread Type (Brot):")
print(df.groupby("Brot", observed=True)["Ameisen"].std())

print("\nBy Topping (Belag):")
print(df.groupby("Belag", observed=True)["Ameisen"].std())

print("\nBy Butter Presence (Butter):")
print(df.groupby("Butter", observed=True)["Ameisen"].std())
```

```python
from scipy.stats import shapiro, levene
import statsmodels.api as sm
import matplotlib.pyplot as plt


# Fit the ANOVA model
model = ols('Ameisen ~ C(Brot) + C(Belag) + C(Butter)', data=df).fit()
residuals = model.resid

# --- Normality test (Shapiro-Wilk) ---
shapiro_test = shapiro(residuals)
print("Shapiro-Wilk Test: W=%.3f, p=%.3f" % (shapiro_test.statistic, shapiro_test.pvalue))

# --- Homogeneity of variances (Levene’s test) ---
levene_test = levene(
    df[df['Brot'] == 'Mehrkorn']['Ameisen'],
    df[df['Brot'] == 'Roggen']['Ameisen'],
    df[df['Brot'] == 'Vollkorn']['Ameisen'],
    df[df['Brot'] == 'Weiß']['Ameisen']
)
print("Levene’s Test: W=%.3f, p=%.3f" % (levene_test.statistic, levene_test.pvalue))

# --- QQ plot for residuals ---
sm.qqplot(residuals, line='45')
plt.title("Q-Q plot of ANOVA residuals")
plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Build model (main effects only)
model = ols('Ameisen ~ C(Brot) + C(Belag) + C(Butter)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
from statsmodels.stats.multitest import multipletests

pvals = anova_table["PR(>F)"]

reject, pvals_corrected, _, _ = multipletests(pvals, alpha=0.05, method='holm')

anova_table["p_corrected"] = pvals_corrected
anova_table["reject_null"] = reject
anova_table
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Example: if 'Brot' was significant
tukey_brot = pairwise_tukeyhsd(endog=df['Ameisen'],
                               groups=df['Brot'],
                               alpha=0.05)
print(tukey_brot)

# For 'Belag'
tukey_belag = pairwise_tukeyhsd(endog=df['Ameisen'],
                                groups=df['Belag'],
                                alpha=0.05)
print(tukey_belag)

# --- Tukey HSD: Butter presence ---
tukey_butter = pairwise_tukeyhsd(endog=df['Ameisen'],
                                 groups=df['Butter'],
                                 alpha=0.05)
print(tukey_butter)
```
